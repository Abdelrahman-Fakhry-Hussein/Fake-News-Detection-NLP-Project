## 3- Text Preprocessing:

```python

def PreProcess(text):
    # Convert all characters to lowercase
    text = text.lower()
    
    # Remove text within square brackets
    text = re.sub('\[.*?\]', '', text)
    
    # Replace any non-word characters (like punctuation) with a space
    text = re.sub("\\W", " ", text)
    
    # Remove URLs starting with 'http://' or 'https://' and URLs starting with 'www.'
    text = re.sub('https?://\S+|www\.\S+', '', text)
    
    # Remove any HTML tags
    text = re.sub('<.*?>+', '', text)
    
    # Remove any punctuation characters
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    
    # Remove any newline characters
    text = re.sub('\n', '', text)
    
    # Remove any words containing digits
    text = re.sub('\w*\d\w*', '', text)
    
    # Return the processed text
    return text
```

```python

#pre processing steps like lower case, stemming and lemmatization

stop = stopwords.words('english')
News['title'] = News['title'].apply(PreProcess)
News['text'] = News['text'].apply(PreProcess)


News['title'] = News['title'].apply(lambda x: " ".join(x.lower() for x in x.split()))
News['title'] = News['title'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
lemmatizer = WordNetLemmatizer()

News['title'] = News['title'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))


News['text'] = News['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
News['text'] = News['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

News['text'] = News['text'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))

News.head()
```

## 4- Feature Extraction & Data Preparation:

```python
#Splitting data into train and validation
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(News[['title','text']], News['label'])


vectorX1 = TfidfVectorizer()
vectorX2 = TfidfVectorizer()
def computeTF_IDF(trainX1, trainX2, testX1, testX2):
    
    tfidfX1 = vectorX1.fit_transform(trainX1)
    tfidfX2 = vectorX2.fit_transform(trainX2)

    tfidf_testX1= vectorX1.transform(testX1)
    tfidf_testX2 = vectorX2.transform(testX2)
 
    # Optionally, you can insert code from * to ** here from below.
    # Stack matrices horizontally (column wise) using hstack().
    trainX_tfidf = scipy.sparse.hstack([tfidfX1, tfidfX2])
    testX_tfidf = scipy.sparse.hstack([tfidf_testX1, tfidf_testX2])

    # Convert Spare Matrix into an Array using toarray() 
    trainX_tfidf_arr = trainX_tfidf.toarray()
    testX_tfidf_arr = testX_tfidf.toarray()
    
    return trainX_tfidf_arr,testX_tfidf_arr


xtrain_tfidf, xvalid_tfidf = computeTF_IDF(train_x['title'],train_x['text'],valid_x['title'],valid_x['text'])

```

## 5-Model Training:

```python
import pickle
import os 
def train_model(classifier, feature_vector_train, label, feature_vector_valid, Y_val, model_filename):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label)
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)

    cm1 = confusion_matrix(Y_val, predictions)

    print('Confusion Matrix for PassiveAggressiveClassifier')
    plot_confusion_matrix(conf_mat=cm1, show_absolute=True,
                          show_normed=True,
                          colorbar=True, class_names=['FAKE', 'REAL'])

    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(model_filename), exist_ok=True)
    # Save the trained model using pickle
    with open(model_filename, 'wb') as f:
        pickle.dump(classifier, f)

    return accuracy_score(predictions, Y_val)
```
